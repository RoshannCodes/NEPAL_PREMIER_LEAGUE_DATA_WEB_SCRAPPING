```python
def clean(text):
    if isinstance(text,str):
        if '-' in text:
            return 0
        text= text.split('*')[0] if '*' in text else text
        return int(text)
    return int(text)
```

```python
df['HS'] = df['HS'].apply(clean)
```

### Key Findings
Players with higher individual scores, better averages, and more runs tend to have better strike rates.

Aggressive playstyles, indicated by hitting more boundaries and sixes, also positively impact the strike rate.

Features like the number of matches, half-centuries, or centuries have limited direct influence on a player’s strike rate.

The occurrence of ducks (zero scores) has a minor negative effect on the strike rate but isn’t a major determining factor.

```python
teams = df['Team Name'].unique()

for team in teams:
    team_data = df[df['Team Name'] == team]
    top_3_batsmen = team_data.sort_values(by='Runs', ascending=False).head(3)
    team_total_runs = team_data['Runs'].sum()
    top_3_runs = list(top_3_batsmen['Runs'])
    other_runs = team_total_runs - sum(top_3_runs)
    
    labels = list(top_3_batsmen['Player']) + ['Other Players']
    values = top_3_runs + [other_runs]
    
    plt.figure(figsize=(4, 4))
    plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, 
            colors=['gold', 'lightblue', 'lightgreen', 'grey'])
    plt.title(f"Proportion of Runs Scored by Top 3 Batsmen for {team}")
    plt.show()
```

```python
df['NO%'] = df['NO'] / df['Inns']
w = [0.4,0.4,0.2]
df['Under Pressure Performance'] = w[0]*df['SR'] + w[1] * df['Ave'] + w[2] * df['NO%']
df.head()
```

```python
top_performer = df.sort_values(by='Under Pressure Performance',ascending=False) 
top_performer
```

```python
top_performer = top_performer.iloc[0]

print(f"🔥 TOP PERFORMER ALERT 🔥\n")
print(f"🏏 Player: {top_performer['Player']} ")
print(f"⚡ Team: {top_performer['Team Name']} ")
print(f"💥 Strike Rate: {top_performer['SR']:.2f}")
print(f"📊 Average: {top_performer['Ave']:.2f}")
print(f"🏃 Not Out Percentage: {top_performer['NO%']:.2f}%")
```
